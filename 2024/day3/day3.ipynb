## Part 1

```python
import re
```

```python
sum = 0
with open('../../input/2024/day3/input.txt', 'r') as f:
    for line in f.readlines():
        l = line.strip()
        #print(l)
        res = re.findall(r"mul\([1-9][0-9]{0,2},[1-9][0-9]{0,2}\)", l)
        
        if res:
            #print(res)
            for s in res:
                s1 = s.split(',')
                sum += int(s1[0][4:])* int(s1[1][:-1])
print(sum)
```

## Part 2

```python
sum = 0
mulActive = True
with open('../../input/2024/day3/input.txt', 'r') as f:
    for line in f.readlines():
        l = line.strip()
        #print(l)
        res = re.findall(r"(mul\([1-9][0-9]{0,2},[1-9][0-9]{0,2}\))|(do\(\))|(don't\(\))", l)
        
        if res:
            print(res)
            for s in res:
                if s[0]: # multiplication
                    print("s[0]", s[0])
                    if mulActive:
                        s1 = s[0].split(',')
                        sum += int(s1[0][4:])* int(s1[1][:-1])
                if s[1]: # don't
                    print("s[1]", s[1])
                    if not mulActive:
                        mulActive = True
                        print("enabling mul")
                if s[2]: # do()
                    print("s[2]", s[2])
                    if mulActive:
                        mulActive = False
                        print("disabling mul")
                    
               
print(sum)
```
