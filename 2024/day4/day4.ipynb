## Part 1

```python
import re
```

```python
def searchXmas(line):
    res1 = re.findall(r"XMAS", line)
    res2 = re.findall(r"SAMX", line)
    return len(res1) + len(res2)
    
```

```python
horizontal = []
vertical = []
diagUp = [] #left right, bottom up
diagDown = [] # right left, up bottom
with open('../../input/2024/day4/input.txt', 'r') as f:
    j = 0
    for line in f.readlines():
        l = line.strip()
        # horizonal
        horizontal.append(l)

        # vertical
        if not vertical:
            for letter in l:
                vertical.append(letter)
        else:
            for i in range(0, len(l)):
                vertical[i] += l[i]

        # diagUp
        if not diagUp:
            for letter in l:
                diagUp.append(letter)
        else:
            for i in range(0, len(l)):
                idx = j + i
                if len(diagUp) <= idx:
                    diagUp.append(l[i])
                else:
                    diagUp[idx] = l[i] + diagUp[idx]

        # diagDown
        if not diagDown:
            for letter in l:
                diagDown.insert(0, letter)
        else:
            for i in range(0, len(l)):
                idx = len(l)-1 +j - i
                if len(diagDown) <= idx:
                    diagDown.append(l[i])
                else:
                    diagDown[idx] = l[i] + diagDown[idx]
        #print(searchXmas(l))
        j = j + 1
print("horizontal", horizontal)
print("vertical", vertical)
print("diagUp", diagUp)
print("diagDown", diagDown)
total = 0
for l in horizontal:
    total += searchXmas(l)
for l in vertical:
    total += searchXmas(l)
for l in diagUp:
    total += searchXmas(l)
for l in diagDown:
    total += searchXmas(l)
print(total)
```
