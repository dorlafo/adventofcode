## Part 1

```python
def predict_next_value(values: []):
    print(values)
    differences = [values[:]]
    cond = False
    while not cond:
        i = len(differences)
        differences.append([])
        cond = True
        for j in range(0, len(differences[i-1])-1):
            diff = differences[i-1][j+1]-differences[i-1][j]
            if diff != 0:
                cond = False
            differences[i].append(diff)
        print(differences[i])
    for i in range(len(differences)-2, -1, -1):
        differences[i].append(differences[i][-1] + differences[i+1][-1])
    print(differences)
    return differences[0][-1]
def solve(file):
    res = 0
    with open(file) as f:
        for line in f:
            values = [int(x.strip()) for x in line.split(' ')]
            
            res += predict_next_value(values)
            print("")
    return res
```

```python
print(solve('input.txt'))
```

```python
def predict_next_value(values: []):
    print(values)
    differences = [values[:]]
    cond = False
    while not cond:
        i = len(differences)
        differences.append([])
        cond = True
        for j in range(0, len(differences[i-1])-1):
            diff = differences[i-1][j+1]-differences[i-1][j]
            if diff != 0:
                cond = False
            differences[i].append(diff)
        print(differences[i])
        
    for i in range(len(differences)-2, -1, -1):
        differences[i].insert(0, differences[i][0] - differences[i+1][0])
    print(differences)
    return differences[0][0]

def solve(file):
    res = 0
    with open(file) as f:
        for line in f:
            values = [int(x.strip()) for x in line.split(' ')]
            
            res += predict_next_value(values)
            print("")
    return res
```

```python
print(solve("input.txt"))
```
