## Part 1

```python
seeds = []
to_soil = []
to_fertilizer = []
to_water = []
to_light = []
to_temperature = []
to_humidity = []
to_location = []
mode = ""
with open('input.txt') as f:
    for line in f:
        if line.startswith("seeds: "):
            seeds = [int(n) for n in line.strip()[len("seeds: "):].split(" ")]

        if line.startswith("seed-to-soil"):
            mode = "to_soil"
        if line.startswith("soil-to-fertilizer"):
            mode = "to_fertilizer"
        if line.startswith("fertilizer-to-water"):
            mode = "to_water"
        if line.startswith("water-to-light"):
            mode = "to_light"
        if line.startswith("light-to-temperature"):
            mode = "to_temperature"
        if line.startswith("temperature-to-humidity"):
            mode = "to_humidity"
        if line.startswith("humidity-to-location"):
            mode = "to_location"
            
        if line[0].isdigit():
            numbers = [int(n) for n in line.strip().split(' ')]
            if (mode == "to_soil"):
                to_soil.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
            elif (mode == "to_fertilizer"):
                to_fertilizer.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
            elif (mode == "to_water"):
                to_water.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
            elif (mode == "to_light"):
                to_light.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
            elif (mode == "to_temperature"):
                to_temperature.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
            elif (mode == "to_humidity"):
                to_humidity.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
            elif (mode == "to_location"):
                to_location.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
                
#print(to_soil)
#print(to_fertilizer)
#print(to_water)
#print(to_light)
#print(to_temperature)
#print(to_humidity)
#print(to_location)

def process(seed: int, to_fcts):
    for to_fct in to_fcts:
        if seed >= to_fct['src_range_start'] and seed <= to_fct['src_range_start'] + to_fct['range']:
            return to_fct['dest_range_start'] + (seed - to_fct['src_range_start'])
    return seed

dest_seed = []
for seed in seeds:
    s = seed
    for to_fct in [to_soil, to_fertilizer, to_water, to_light, to_temperature, to_humidity, to_location]:
        s  = process(s, to_fct)
    dest_seed.append(s)
print(min(dest_seed))
```
