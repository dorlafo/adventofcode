```python
seeds = []
to_soil = []
to_fertilizer = []
to_water = []
to_light = []
to_temperature = []
to_humidity = []
to_location = []
mode = ""
with open('input.txt') as f:
    for line in f:
        if line.startswith("seeds: "):
            seeds = [int(n) for n in line.strip()[len("seeds: "):].split(" ")]

        if line.startswith("seed-to-soil"):
            mode = "to_soil"
        if line.startswith("soil-to-fertilizer"):
            mode = "to_fertilizer"
        if line.startswith("fertilizer-to-water"):
            mode = "to_water"
        if line.startswith("water-to-light"):
            mode = "to_light"
        if line.startswith("light-to-temperature"):
            mode = "to_temperature"
        if line.startswith("temperature-to-humidity"):
            mode = "to_humidity"
        if line.startswith("humidity-to-location"):
            mode = "to_location"
            
        if line[0].isdigit():
            numbers = [int(n) for n in line.strip().split(' ')]
            if (mode == "to_soil"):
                to_soil.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
            elif (mode == "to_fertilizer"):
                to_fertilizer.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
            elif (mode == "to_water"):
                to_water.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
            elif (mode == "to_light"):
                to_light.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
            elif (mode == "to_temperature"):
                to_temperature.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
            elif (mode == "to_humidity"):
                to_humidity.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
            elif (mode == "to_location"):
                to_location.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2]})
                
#print(to_soil)
#print(to_fertilizer)
#print(to_water)
#print(to_light)
#print(to_temperature)
#print(to_humidity)
#print(to_location)

def process(seed: int, to_fcts):
    for to_fct in to_fcts:
        if seed >= to_fct['src_range_start'] and seed <= to_fct['src_range_start'] + to_fct['range']:
            return to_fct['dest_range_start'] + (seed - to_fct['src_range_start'])
    return seed

dest_seed = []
for seed in seeds:
    s = seed
    for to_fct in [to_soil, to_fertilizer, to_water, to_light, to_temperature, to_humidity, to_location]:
        s  = process(s, to_fct)
    dest_seed.append(s)
print(min(dest_seed))
```

```python
seeds = []
to_soil = []
to_fertilizer = []
to_water = []
to_light = []
to_temperature = []
to_humidity = []
to_location = []
mode = ""
with open('example.txt') as f:
    for line in f:
        if line.startswith("seeds: "):
            seeds_range = [int(n) for n in line.strip()[len("seeds: "):].split(" ")]

            for i in range(0, len(seeds_range), 2):
                seeds.append({"start": seeds_range[i], "end": seeds_range[i]+seeds_range[i+1]-1})
        if line.startswith("seed-to-soil"):
            mode = "to_soil"
        if line.startswith("soil-to-fertilizer"):
            mode = "to_fertilizer"
        if line.startswith("fertilizer-to-water"):
            mode = "to_water"
        if line.startswith("water-to-light"):
            mode = "to_light"
        if line.startswith("light-to-temperature"):
            mode = "to_temperature"
        if line.startswith("temperature-to-humidity"):
            mode = "to_humidity"
        if line.startswith("humidity-to-location"):
            mode = "to_location"
            
        if line[0].isdigit():
            numbers = [int(n) for n in line.strip().split(' ')]
            if (mode == "to_soil"):
                to_soil.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2], "name":"to_soil"})
            elif (mode == "to_fertilizer"):
                to_fertilizer.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2], "name":"to_fertilizer"})
            elif (mode == "to_water"):
                to_water.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2], "name":"to_water"})
            elif (mode == "to_light"):
                to_light.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2], "name":"to_light"})
            elif (mode == "to_temperature"):
                to_temperature.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2], "name":"to_temperature"})
            elif (mode == "to_humidity"):
                to_humidity.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2], "name":"to_humidity"})
            elif (mode == "to_location"):
                to_location.append({"dest_range_start":numbers[0], "src_range_start":numbers[1], "range":numbers[2], "name":"to_location"})
                
#print(to_soil)
#print(to_fertilizer)
#print(to_water)
#print(to_light)
#print(to_temperature)
#print(to_humidity)
#print(to_location)
#print("seeds:", seeds)

def process(seeds, to_fcts):
    res = []
    for to_fct in to_fcts:
        for seed in seeds:
            seed_start = seed['start']
            seed_end = seed['end']
            src_start = to_fct['src_range_start']
            src_end = to_fct['src_range_start'] + to_fct['range'] - 1
            dst_start = to_fct['dest_range_start']
            dst_end = to_fct['dest_range_start'] + to_fct['range'] - 1
            
            print("Trying to map seed [{},{}] with function {} and mapping : [{},{}] -> [{},{}]".format(seed_start, seed_end, to_fct['name'], src_start, src_end, dst_start, dst_end))

            map_start = max(src_start, seed_start)
            map_end = min(src_end, seed_end)

            if map_end >= map_start:
                if map_start != seed_start:
                    before_start = min(src_start, seed_start)
                    before_end = map_start-1
                else:
                    before_start = before_end = 0
                if map_end != seed_end:
                    after_start = map_end + 1
                    after_end = max(src_end, seed_end)
                else:
                    after_start = after_end = 0
                    
                mapped_start = dst_start + (map_start-src_start)
                mapped_end = mapped_start + map_end - map_start
                print("Mapping found:")
                print("mapping : [{},{}] -> [{},{}]".format(src_start, src_end, dst_start, dst_end))
                print("map : [{},{}]u[{},{}]u[{},{}]".format(before_start or "", before_end or "", map_start, map_end, after_start or "", after_end or ""))
                print("[{},{}] mapped to [{},{}]".format(map_start, map_end, mapped_start, mapped_end))
                
                if before_start != before_end != 0:
                    print("ret call")
                    r = process([{'start':before_start, 'end':before_end}], to_fcts)
                    for e in r:
                        res.append(e)
                    
                res.append({'start':mapped_start, 'end':mapped_end})
                
                
                if after_start != after_end != 0:
                    print("ret call")
                    r = process([{'start':after_start, 'end':after_end}], to_fcts)
                    for e in r:
                        res.append(e)
            else:
                print("No mapping found.")
    if res != []:
        return res
    else:
        return seeds
print("Seeds:", seeds)
dest_seed = []
for seed_r in seeds:
    print("Processing seed [{},{}]".format(seed_r['start'], seed_r['end']))
    s = [seed_r]
    for to_fct in [to_soil, to_fertilizer, to_water, to_light, to_temperature, to_humidity, to_location]:
        #print("initial s : {} with fct {}".format(s, to_fct[0]["name"]))
        s  = process(s, to_fct)
        print("returned ranges :", s)
        print("")
    print("------------------------")
    for e in s:    
        dest_seed.append(e)
min_dest = []
for ds in dest_seed:
    min_dest.append(min(ds['start'], ds['end']))
print(min(min_dest))
```

Problème actuel : si une partie est déjà mappée par une range d'une fonction, il faut enlever cette partie avant de tester sur les autres fonctions, sinon cela ajoute des valeurs déjà mappée dans les unions

example:

No mapping of [74,87] with function to_temperature and mapping : [45,63] -> [81,99]
