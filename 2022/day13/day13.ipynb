```python
"""
result of compare: 0 = same, 1 = right order, -1 = not right order
"""
indent = '  '
def compare(left, right, depth=0):
    print("{}- Compare {} vs {}".format(depth*indent, left, right))
    if type(left) == type(right):
        if type(left) == int:
            if left < right:
                # right order
                print("{}- Left side is smaller, so inputs are in the right order".format(depth*indent))
                return 1
            elif left > right:
                # not in right order
                print("{}- Right side is smaller, so inputs are not in the right order".format(depth*indent))
                return -1
            else:
                # same integer, check next
                return 0
        else:
            for index, e in enumerate(left):
                if index < len(right):
                    res = compare(e, right[index], depth+1)
                    if res != 0:
                        return res
                else:
                    print("{}- Right side ran out of items, so inputs are not in the right order".format(depth*indent))
                    return -1
            if len(left) < len(right):
                print("{}- Left side ran out of items, so inputs are in the right order".format(depth*indent))
                return 1
            else:
                return 0
            
                    
    else:
        if type(left) == int:
            print("{}- Mixed types; convert left to [{}] and retry comparison".format(depth*indent, left))
            return compare([left], right, depth+1)
        else:
            print("{}- Mixed types; convert right to [{}] and retry comparison".format(depth*indent, right))
            return compare(left, [right], depth+1)
```

## Part 1

```python
pairs = [[]]
idx = 0
with open('input.txt', encoding="utf-8") as f:
    for line in f:
        if line == '\n':
            idx += 1
            pairs.append([])
        else:
            pairs[idx].append(eval(line.strip('\n')))

idx = 1
sum_in_order = 0
for left, right in pairs:
    print("== Pair {} ==".format(idx))
    order = compare(left, right)
    if order == 1:
        sum_in_order += idx
    print("Is in right order:", order)
    print()
    idx += 1
print(sum_in_order)
```
