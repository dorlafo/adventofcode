```python
def parse(l: str):
    print("parsing", l)
    id_start = l.rfind('[')
    id_end = l.find(']', id_start)
    if id_start != -1 and id_end != -1:
        l1 = l[:id_start].strip(',')
        l2 = l[id_start+1: id_end].strip(',')
        l3 = l[id_end+1:].strip(',')
        print(l)
        print("-'{}'".format(l1))
        print("-'{}'".format(l2))
        print("-'{}'".format(l3))
        res = []
        if len(l1) != 0:
            res.append(parse(l1))
        if len(l2) != 0:
            res.append(parse(l2))
        if len(l3) != 0:
            res.append(parse(l3))
        
        if len(res) == 1:
            res = res[0]
        print("return", res)
        return res
    else:
        res = [int(e) for e in l.split(',') if e.isdigit()]
        print("return", res)
        return res
```

```python
pairs = [[]]
idx = 0
with open('input_test.txt', encoding="utf-8") as f:
    for line in f:
        if line == '\n':
            idx += 1
            pairs.append([])
        else:
            pairs[idx].append(parse(line.strip('\n')))

print("pairs")
for p in pairs:
    print(p[0], '--', p[1])
```
