```python
fs = {'/' : {}}
with open('input.txt', encoding="utf-8") as f:
    curr_dir = []
    ptr = {}
    nb_file = 0
    nb_dir = 0
    for line in f:
        cmd = line.strip('\n').split(' ')
        #print("cmd =",cmd)
        if cmd[0] == '$':
            if cmd[1] == 'cd':
                if cmd[2] == '/':
                    curr_dir = ['/']
                elif cmd[2] == '..':
                    curr_dir.pop()
                else:
                    curr_dir.append(cmd[2])
                print("change dir to", curr_dir)
            else:
                # ls case, point to current dir in map
                ptr = fs[curr_dir[0]]
                for d in curr_dir[1:]:
                    ptr = ptr[d]
        elif cmd[0] == 'dir':
            nb_dir += 1
            dir_name = cmd[1]
            ptr[dir_name] = {}
        else:
            nb_file += 1
            file_size = cmd[0]
            file_name = cmd[1]
            ptr[file_name] = {'size': file_size}
    print("file:", nb_file)
    print("dir:", nb_dir)
#print("fs = ", fs)
```

```python
def print_fs(fs, depth):
    placeholder = '  '
    for k,v in fs.items():
        if 'size' in v:
            print("{}- {} (file, size={})".format(depth*placeholder, k, v['size']))
            #print(depth*'   ', '- ', k, '(file, size =', v['size'], ')')
        else:
            print("{}- {} (dir)".format(depth*placeholder, k))
            #print(depth*'   ', '- {} (dir)'.format(k))
            print_fs(v, depth+1)
```

```python
print_fs(fs, 0)
```

```python
def get_dir_size(d):
    size = 0
    dir_map = {}
    for k,v in d.items():
        if 'size' in v:
            size += int(v['size'])
        else:
            dir_size, tmp_map = get_dir_size(v)
            if k in dir_map:
                dir_map[k+str(dir_size)] = dir_size
            else:
                dir_map[k] = dir_size            
            print('dir {} has size {}'.format(k, dir_size))
            for k,v in tmp_map.items():
                if k in dir_map:
                    dir_map[k+str(v)] = v
                else:
                    dir_map[k] = v
                
                
            size += dir_size
            
    return size, dir_map
        
```
